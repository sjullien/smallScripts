```python
import torch
import copy

class Individual:
    
    def __init__(self):
        self.genes = torch.randint(0,2,(10,))
        self.fitness = torch.sum(self.genes)
        
class Population:
    
    def __init__(self,pop_size):
        self.individuals = []
        self.pop = pop_size
        for i in range(0, pop_size):
            self.individuals.append(Individual())
        self.fittest = self.individuals[self.getFittest()]
        self.secondFittest = self.individuals[self.getSecondFittest()]
        self.weakest = self.individuals[self.getWeakest()]

    
    def getFittest(self):
        max_index = 0
        for i in range(0, self.pop):
            if self.individuals[i].fitness > self.individuals[max_index].fitness :
                max_index = i
        return(max_index)
                
    def getWeakest(self):
        min_index = 0
        for i in range(0, self.pop):
            if self.individuals[i].fitness < self.individuals[min_index].fitness :
                min_index = i
        return(min_index)
    
    def getSecondFittest(self):
        max_index = self.getFittest()
        max_2 = 0
        for i in range(0, self.pop):
            if self.individuals[i].fitness > self.individuals[max_2].fitness  and i != max_index:
                max_2 = i
        return(max_2)
    
    def getFittestChildren(self):
        if self.fittest.fitness > self.secondFittest.fitness:
            return self.fittest
        else:
            return self.secondFittest
    
    def addFittestChildren(self):
        min_index = self.getWeakest()
        self.individuals[min_index] = copy.deepcopy(self.fittest)
    
    def update(self):
        self.fittest = self.individuals[self.getFittest()]
        self.secondFittest = self.individuals[self.getSecondFittest()]
        self.weakest = self.individuals[self.getWeakest()]
                                        
    def crossover(self):
            crossIndex = int(torch.randint(0, len(self.individuals[0].genes),(1,)))
            for i in range(0, crossIndex):
                temp = self.fittest.genes[i]
                self.fittest.genes[i] = self.secondFittest.genes[i]
                self.secondFittest.genes[i] = temp
            self.secondFittest.fitness = torch.sum(self.secondFittest.genes)
            self.fittest.fitness = torch.sum(self.fittest.genes)
    
    def mutation(self):
        mutationIndex = int(torch.randint(0, len(self.individuals[0].genes),(1,)))
        if self.fittest.genes[mutationIndex] == 0:
            self.fittest.genes[mutationIndex] = 1
        else:
            self.fittest.genes[mutationIndex] = 0
            
        mutationIndex = int(torch.randint(0, len(self.individuals[0].genes),(1,)))
        if self.secondFittest.genes[mutationIndex] == 0:
            self.secondFittest.genes[mutationIndex] = 1
        else:
            self.secondFittest.genes[mutationIndex] = 0
            
            
    
```

```python
pop = Population(10)
generation = 0

while pop.fittest.fitness < 10:
    print('Generation : '+ str(generation) +', fittest: '+str(pop.fittest.fitness))
    pop.crossover()
    if  int(torch.randint(0,10,(1,)))<5:
        pop.mutation()
    pop.addFittestChildren()
    pop.update()
    generation+=1
for i in range(0,10):
    print(pop.individuals[i].genes)
```
